```python
#file path
ROOT_DIR = os.getcwd()
DATA_DIR = os.path.join(ROOT_DIR, 'data')
DATA_DIR
```

```python
#df is the name of the table that holds your data after you load it in Python
dta_file = os.path.join(DATA_DIR, 'tographinFigure6.dta')
df = pd.read_stata(dta_file)
df.head(7)
```

```python
#replace age=age+1
df['age'] = df['age'] + 1
```

```python
df.head(7)
```

```python
plt.figure(figsize=(6,4))

# plot each line like Stata's "connect"
plt.plot(df['age'], df['ap'],          linewidth=1, color='black', marker='o', markersize=3, label="No Access")
plt.plot(df['age'], df['pilllegal'],   linewidth=1, color='red', marker='o', markersize=3, label="Pill legal, cannot consent")
plt.plot(df['age'], df['pillconsent'], linewidth=1, color='orange', marker='o', markersize=3, label="Pill legal, can consent")
plt.plot(df['age'], df['abortlegal'],  linewidth=1, color='blue', marker='o', markersize=3, label="Abortion legal, cannot consent")
plt.plot(df['age'], df['abortconsent'],linewidth=1, color='green', marker='o', markersize=3, label="Abortion legal, can consent")

# formatting to match Stata
plt.xlabel("Age")
plt.ylabel("Adjusted predicted probability")

plt.yticks([0, .10, .20, .30, .40])

plt.legend(
    loc='upper center', bbox_to_anchor=(0.5, -0.25), ncol=3, fontsize=8, frameon=False)
plt.tight_layout()

# Export
plt.savefig("Results/Figure6.pdf")
plt.savefig("Results/Figure6.eps")
plt.show()
```
